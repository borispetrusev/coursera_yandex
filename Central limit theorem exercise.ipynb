# Central limit theorem exercise

Let's import libraries that we need

```python
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

I decided to pick a gamma distribution for this exercise

Let's pick k=3 and Theta = 1

Generating sample of 1000

```python
k=3
Theta = 1
sample = gamma.rvs(a=k,scale=Theta,size=1000)
```

Plotting a histogram of sample distribution and theoretical gamma pdf

```python
plt.hist(sample, normed=True)
x = np.linspace(gamma.ppf(0.01, k),gamma.ppf(0.99, k), 100)
plt.plot(x, gamma.pdf(x, k),'r-', lw=5, alpha=0.6, label='theoretical gamma pdf')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```

Let's create a function for generating means

```python
def generate_means(n,sample_size):
    means=[]
    for i in range(sample_size):
        sum = 0.0
        for element in np.random.gamma(k, Theta, size=n):
            sum+=element
        mean = sum/n
        means.append(mean)
    return means
```
